## Get unique values of a column

## Search for data

```python
movies[movies.duration > 120].head()
```

```python
movies[movies.genre == 'Action'].head()
```

## Restrict columns after search

```python
movies.loc[movies.genre == 'Action', ['title', 'star_rating', 'duration']].head()
```

## Group data and show statistical info

```python
movies.groupby(['genre'])['duration'].mean() # get average duration for each genre
```

## Data visualization

```python
%matplotlib inline
```

```python
movies.groupby('genre').size().plot.bar() # Show movies produced by each genre
```

```python
movies.groupby('genre')['duration'].mean().plot.bar() # Show average duration of each genre
```

## Groupby method returns a dictionary of DataFrames

```python
for genre, df in movies.groupby('genre'):
    print(genre, type(df))
```

## Data manipulation: merge (concat)

```python
df = pd.DataFrame(np.random.rand(10,4), columns=list('abcd'))
df
```

```python
pieces = [df[:5], df[2:7], df[6:]]
pd.concat(pieces).head(10)  # concat merges the data by key
```
